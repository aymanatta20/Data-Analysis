```python
from  collections import Counter
import pandas as pd
```

```python
x = df.Brand
count = Counter(x)
df=pd.DataFrame.from_dict(count ,orient ="index")
df[0].sort_values(ascending=True)[:100]
df.plot(kind="bar")
plt.title("most requested model")
plt.xlabel("Brand")
plt.ylabel("quantities")
plt.show()
```
