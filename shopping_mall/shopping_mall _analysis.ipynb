```python
quarter =  [ ]

for i in data.month: 
    if i <= 3: 
        quarter.append(1)
    elif i <= 6: 
        quarter.append(2)
    elif i <= 9: 
        quarter.append(3)
    else:
        quarter.append(4)
    
```

```python
data["quarter"] = quarter
```

```python
data 
```

```python
pd.crosstab([data.year, data.quarter] , data.shopping_mall , values= data.price , aggfunc= "sum")
```
